# **Q1**

```python
from huggingface_hub import InferenceClient
from getpass import getpass
```

```python
HF_TOKEN = getpass("Enter your HF token: ")

client = InferenceClient(
    token=HF_TOKEN,
    model="deepset/roberta-base-squad2"
)
```

```python
context = "The James Webb Space Telescope (JWST) is the most powerful space telescope ever built. It was launched by NASA in December 2021 and is designed to observe the most distant galaxies, stars, and planets in the universe. Unlike the Hubble Space Telescope, JWST uses infrared technology, allowing it to peer through dust clouds and capture images of objects that are billions of light-years away. This capability helps scientists understand how galaxies form, how stars are born, and whether there are signs of life on exoplanets."
question = "What makes the James Webb Space Telescope different from the Hubble Space Telescope?"
```

```python
try:
    # This line sends the question and context to the model
    output = client.question_answering(question=question, context=context)

    # This line prints the result if successful
    print("Answer:", output.answer)  # Only print the text answer

except Exception as e:
    # If something goes wrong (like API error, wrong model, token error), this will run
    print("Error occurred:", e)
```

# **Q2**

```python
# Import necessary libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
def generate_story(prompt, temperature=0.7):
    try:
        # Load pre-trained model and tokenizer
        tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
        model = GPT2LMHeadModel.from_pretrained("gpt2")

        # Encode the input prompt
        input_ids = tokenizer.encode(prompt, return_tensors="pt")

        # Generate output using temperature and sampling parameters
        output = model.generate(
            input_ids,
            do_sample=True,
            temperature=temperature,
            top_k=50,
            top_p=0.95,
            max_length=150,
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id  # ✅ This prevents warnings
        )

        # Decode and print the generated text
        story = tokenizer.decode(output[0], skip_special_tokens=True)
        print(f"\n--- Temperature: {temperature} ---\n{story}\n")

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
prompt = "Once upon a time in a futuristic city"
temperatures = [0.3, 0.7, 1.0, 1.3]

for temp in temperatures:
    generate_story(prompt, temperature=temp)
```
